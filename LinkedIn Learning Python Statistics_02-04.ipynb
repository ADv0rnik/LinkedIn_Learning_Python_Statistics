```python
bshort = bb[['artist.inverted', 'track', 'time','date.entered','x1st.week', 'x2nd.week', 'x3rd.week']]
bshort.head()
```

```python
bshort.columns=['artist','track','time','date.entered','wk1', 'wk2', 'wk3']
bshort.head()
```

```python
dmelt = bshort.melt(['artist','track','time','date.entered'],['wk1', 'wk2', 'wk3'],'week','rank')
dmelt.head()
```

```python
dmelt.query('track == "Liar"')
```

```python
dmelt['week'] = dmelt['week'].apply(lambda s: int(s[2]))
```

```python
dmelt['date.entered']=pd.to_datetime(dmelt['date.entered'])
```

```python
dmelt['date'] = dmelt['date.entered']
```

```python
dmelt.drop(['date.entered'])
```

```python
dmelt.drop(['date.entered'], axis = 1, inplace = True)
```

```python
dmelt.head(20)
```

```python
bfinal = dmelt[['artist','track','time', 'date', 'week', 'rank']]
```

```python
bfinal.sort_values(['artist', 'track'], inplace = True)
bfinal.head(20)
```

```python
tracks = bfinal[['artist','track','time']].drop_duplicates()
tracks.head()
```

```python
tracks.index.name ='id'
tracksid = tracks.reset_index()
tracksid.head()
```

```python
tidy = pd.merge(tracksid, bfinal, on =['artist','track','time'])
```
