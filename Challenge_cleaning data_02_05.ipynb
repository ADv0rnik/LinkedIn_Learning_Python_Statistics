```python
import pandas as pd
import numpy as nmp

import matplotlib
import matplotlib.pyplot as plt
```

```python
tb = pd.read_csv('tb.csv')
tb.head(10)
```

```python
tb.columns
```

```python
tbmelt = tb.melt(['country', 'year'],['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],'sexage','cases')
tbmelt.head()
```

```python
tbmelt['sex']=tbmelt['sexage'].str.slice(0,1)
tbmelt['age']=tbmelt['sexage'].str.slice(start = 1)
tbmelt.head()
```

```python
tbmelt.drop(['sexage'], axis =1, inplace = True)
tbmelt.head()
```

```python
tbmelt['age']=tbmelt['age'].map({'04': '0-4', '514':'5-14', '1524':'15-24', '2534':'25-34','3544':'35-44','4554':'45-54','5564':'55-64',
                                '65':'65+', 'u':nmp.nan})
```

```python
tbmelt.head()
```

```python
tbfinal = tbmelt.dropna(subset=['cases'])
tbfinal.head()
```

```python
tbfinal.sort_values(by=['country','year','age','sex']).head()
```

```python
final = tbfinal[['country','year','age','sex','cases']]
final.head()
```

```python
final.to_csv('final1.csv')
```
